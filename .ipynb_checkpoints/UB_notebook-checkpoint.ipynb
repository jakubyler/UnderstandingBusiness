```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('ikano.xlsx', index_col = 0)
df
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns = ['ID', 'gender', 'income', 'marital_status', 'education',
       'contract_type', 'profession', 'sector', 'location', 'age',
       'payment_history', 'supplementary_products',
       'prev_agreements_number', 'credit_bureau_data',
       'purchase_date', 'months_on_borad', 'product',
       'home_type', 'children',
       'household_members', 'home_type_1',
       'rooms', 'area', 'products', 'membership_term',
       'instalment']
```

```python
for col in df.columns[1:]:
    if df[col].dtype == 'object':
        plt.figure(figsize = (10,10))
        sns.countplot(df[col])
        plt.title(col, size = 20)
        plt.xticks(rotation = 90)
        plt.show()
    else:
        plt.figure(figsize = (10,10))
        sns.distplot(df.loc[df[col]<=df[col].quantile(0.99),col], bins = 50)
        plt.title(col, size = 20)
        plt.show()
```

```python
df['time'] = df['purchase_date'].apply(lambda x: int(str(x)[0:4])*12 + int(str(x)[5:7]))
df['time'] = df['time'] - df['time'].min()
df['month'] = df['purchase_date'].apply(lambda x: str(x)[5:7])
df['date'] = df['purchase_date'].apply(lambda x: str(x)[0:7])
```

```python
df.isna().sum()
```

```python
# jak ustalimy która zmienna nas interesuje to możemy jakoś przeanalizować, ale imo żadna strata je wywalić
df = df[~df['education'].isna()].copy() # mała strata, trudno stwierdzić gdzie przypisać
```

```python
df['education'].value_counts()
```

```python
df['profession'].value_counts()
```

```python
df['sector'].value_counts()
```

```python
df['profession'] = df['profession'].fillna('other') # nowa klasa, relatywnie liczna
df['sector'] = df['sector'].fillna('other') # nowa klasa, relatywnie liczna
df['contract_type'] = df['contract_type'].fillna('Full time') # tylko 4 missingi, Full time to moda i jest jej zdecydowana większość
df['payment_history'] = df['payment_history'].fillna(0) # binarna czy ma, nie wiemy czy ma więc konserwatywnie założymy, że nie

df['credit_bureau_data_missing'] = (df['credit_bureau_data'].isna())*1
df['credit_bureau_data'] = df['credit_bureau_data'].fillna(0) # tbh nie wiem jak to traktować, jeżeli zrobimy  z tego kategoryczną to zrobimy dodatkową klasę

df['children_missing'] = (df['children'].isna())*1
df['children'] = df['children'].fillna(0) # tutaj moda i binarna po prostu
```
